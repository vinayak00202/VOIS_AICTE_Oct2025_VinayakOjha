```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel(r"C:\Users\vamit\Downloads\1730285881-Airbnb_Open_Data.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Clean the price column
df['price'] = df['price'].replace({'$': '', ',': ''}, regex=True).astype(float)

# Set the figure size
plt.figure(figsize=(10, 6))

# Create the box plot
sns.boxplot(x='room type', y='price', data=df)

# Set the title and labels
plt.title('Price Distribution by Room Type', fontsize=16)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Price (in USD)', fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1000)  # Limiting y-axis for better visualization of the boxes
plt.tight_layout()

# Show theÂ plot
plt.show()
```

```python
top_neighborhoods = df['neighbourhood'].value_counts().nlargest(10)

# Set the figure size
plt.figure(figsize=(12, 7))

# Create the bar plot
sns.barplot(x=top_neighborhoods.values, y=top_neighborhoods.index, palette='viridis')

# Set the title and labels
plt.title('Top 10 Neighborhoods by Listing Count', fontsize=16)
plt.xlabel('Number of Listings', fontsize=12)
plt.ylabel('Neighborhood', fontsize=12)
plt.tight_layout()

# Show theÂ plot
plt.show()
```

```python
top_hosts = df['host name'].value_counts().nlargest(10)

# Set the figure size
plt.figure(figsize=(12, 7))

# Create the bar plot
sns.barplot(x=top_hosts.values, y=top_hosts.index, palette='plasma')

# Set the title and labels
plt.title('Top 10 Hosts by Listing Count', fontsize=16)
plt.xlabel('Number of Listings', fontsize=12)
plt.ylabel('Host Name', fontsize=12)
plt.tight_layout()

# Show theÂ plot
plt.show()
```
